```python
import sqlite3
import numpy as np
import pandas as pd
from taiwan_election_2020 import TaiwanElection2020

tw_election = TaiwanElection2020()
presidential = tw_election.generate_presidential()
legislative_regional = tw_election.generate_legislative_regional()
legislative_at_large = tw_election.generate_legislative_at_large()
```

```python
print(presidential.shape)
print(legislative_regional.shape)
print(legislative_at_large.shape)
```

```python
# Creating lookup_table
## parties - id/party
legislative_regional_parties = legislative_regional['party'].unique()
legislative_at_large_parties = legislative_at_large['party'].unique()
all_parties = np.concatenate([legislative_regional_parties, legislative_at_large_parties])
unique_parties = np.unique(all_parties)
unique_parties = np.sort(unique_parties)
party_ids = range(1, unique_parties.size + 1)
parties = pd.DataFrame()
parties['id'] = party_ids
parties['party'] = unique_parties
parties
```

```python
# Creating lookup_table
## candidates - id/party_id/number/name/type
presidential_candidates = presidential.groupby(['number', 'candidate'])
legislative_regional_candidates = legislative_regional.groupby(['party', 'number', 'candidate'])
presidential_candidates_numbers = [n for n, p in presidential_candidates.count().index]
presidential_candidates_names = [p for n, p in presidential_candidates.count().index]
presidential_candidates_party_ids = [44, 3, 37]
legislative_regional_candidates_parties = [p for p, n, c in legislative_regional_candidates.count().index]
legislative_regional_candidates_numbers = [n for p, n, c in legislative_regional_candidates.count().index]
legislative_regional_candidates_names = [c for p, n, c in legislative_regional_candidates.count().index]
presidential_df = pd.DataFrame()
presidential_df['party_id'] = presidential_candidates_party_ids
presidential_df['number'] = presidential_candidates_numbers
presidential_df['candidate'] = presidential_candidates_names
presidential_df['type'] = '總統副總統'
legislative_regional_candidates_party_ids = []
for p in legislative_regional_candidates_parties:
    partial_df = parties[parties['party'] == p]
    legislative_regional_candidates_party_ids.append(partial_df['id'].values[0])
legislative_regional_df = pd.DataFrame()
legislative_regional_df['party_id'] = legislative_regional_candidates_party_ids
legislative_regional_df['number'] = legislative_regional_candidates_numbers
legislative_regional_df['candidate'] = legislative_regional_candidates_names
legislative_regional_df['type'] = '區域立委'
candidates = pd.concat([presidential_df, legislative_regional_df])
candidates = candidates.reset_index(drop=True)
candidates['id'] = range(1, candidates.shape[0] + 1)
candidates = candidates[['id', 'party_id', 'type', 'number', 'candidate']]
candidates
```

```python
# Creating lookup_table
## admin_regions - id/county/town/village
county_town_village = presidential.groupby(['county', 'town', 'village']).count().index
counties, towns, villages = [], [], []
for c, t, v in county_town_village:
    counties.append(c)
    towns.append(t)
    villages.append(v)
region_ids = range(1, county_town_village.shape[0] + 1)
admin_regions = pd.DataFrame()
admin_regions['id'] = region_ids
admin_regions['county'] = counties
admin_regions['town'] = towns
admin_regions['village'] = villages
admin_regions
```

```python
# Normalize presidential
presidential_admin_regions = pd.merge(presidential, admin_regions, left_on=['county', 'town', 'village'], right_on=['county', 'town', 'village'])
presidential_admin_regions = presidential_admin_regions[['id', 'office', 'candidate', 'votes']]
presidential_admin_regions.columns = ['admin_region_id', 'office', 'candidate', 'votes']
presidential_candidates = pd.merge(presidential_admin_regions, candidates, left_on='candidate', right_on='candidate')
presidential_normalized = presidential_candidates[['admin_region_id', 'office', 'id', 'votes']]
presidential_normalized.columns = ['admin_region_id', 'office', 'candidate_id', 'votes']
presidential_normalized
```

```python
# Normalize legislative_regional
legislative_regional_admin_regions = pd.merge(legislative_regional, admin_regions, left_on=['county', 'town', 'village'], right_on=['county', 'town', 'village'])
legislative_regional_admin_regions = legislative_regional_admin_regions[['id', 'electoral_district', 'office', 'party', 'candidate', 'votes']]
legislative_regional_admin_regions.columns = ['admin_region_id', 'electoral_district', 'office', 'party', 'candidate', 'votes']
candidates_parties = pd.merge(candidates, parties, left_on='party_id', right_on='id')
legislative_regional_candidates = pd.merge(legislative_regional_admin_regions, candidates_parties, left_on=['candidate', 'party'], right_on=['candidate', 'party'])
legislative_regional_normalized = legislative_regional_candidates[['admin_region_id', 'electoral_district', 'office', 'id_x', 'votes']]
legislative_regional_normalized.columns = ['admin_region_id', 'electoral_district', 'office', 'candidate_id', 'votes']
legislative_regional_normalized
```

```python
# Normalize legislative_at_large
legislative_at_large_admin_regions = pd.merge(legislative_at_large, admin_regions, left_on=['county', 'town', 'village'], right_on=['county', 'town', 'village'])
legislative_at_large_admin_regions = legislative_at_large_admin_regions[['id', 'office', 'number', 'party', 'votes']]
legislative_at_large_admin_regions.columns = ['admin_region_id', 'office', 'number', 'party', 'votes']
legislative_at_large_normalized = pd.merge(legislative_at_large_admin_regions, parties, left_on='party', right_on='party')
legislative_at_large_normalized = legislative_at_large_normalized[['admin_region_id', 'office', 'id', 'votes']]
legislative_at_large_normalized.columns = ['admin_region_id', 'office', 'party_id', 'votes']
legislative_at_large_normalized
```

```python
conn = sqlite3.connect('twElection2020.db')
parties.to_sql('parties', conn, index=False)
candidates.to_sql('candidates', conn, index=False)
admin_regions.to_sql('admin_regions', conn, index=False)
presidential_normalized.to_sql('presidential', conn, index=False)
legislative_regional_normalized.to_sql('legislative_regional', conn, index=False)
legislative_at_large_normalized.to_sql('legislative_at_large', conn, index=False)
```
